**Q.1** Group the given data into four categories for Bright and Dark traits. Where categories are A(1-0),B(3-2), C(6-4) and D(7-10).

For each student calculate the average of bright traits and dark traits and assign them into a category.

**Bright trait**s:  LOGIC, CONCENTRATION, FOCUS, COGNITIVE SKILL, RETENTION POWER, HARD WORKING, STUDY HABIT, CONSCIOUSNESS
**Dark Trait**: SILLY MISTAKES,KNOWLEDGE GAP, IMPULSIVE, LEARNING GAP


```python
categorize = lambda x: 'A' if x <= 1 else 'B' if x <= 3 else 'C' if x <= 6 else 'D'

# Calculate average of dark traits and catogorize
df['Average Dark Traits'] = df[['SILLY MISTAKES', 'KNOWLEDGE GAP', 'IMPULSIVE', 'LEARNING GAP']].mean(axis=1).round()

df['Dark Category'] = df['Average Dark Traits'].apply(categorize)

# Calculate average of bright traits and catogorize
df['Average Bright Traits'] = df[['LOGIC', 'CONCENTRATION', 'FOCUS', 'COGNITIVE SKILL', 'RETENTION POWER',
                                  'HARD WORKING', 'STUDY HABIT', 'CONSCIOUSNESS']].mean(axis=1).round()

# Apply the categorization function to create the 'Bright Category' column
df['Bright Category'] = df['Average Bright Traits'].apply(categorize)

df
```

**Q.2** Write a Python program that counts the frequency of each word in a given text document. Remove any punctuation marks and convert all words to lowercase before counting. Display the **top 5 most frequent words along with their frequencies**. Also calculate the below parameters.

The formulas for calculations are :
**Positive Score:** This score is calculated by assigning the value of +1 for each word if found in the Positive Dictionary and then adding up all the values.
**Negative Score:** This score is calculated by assigning the value of -1 for each word if found in the Negative Dictionary and then adding up all the values. We multiply the score with -1 so that the score is a positive number.

**Polarity Score:** This is the score that determines if a given text is positive or negative in nature. It is calculated by using the formula:
Polarity Score = (Positive Score – Negative Score)/ ((Positive Score + Negative Score) + 0.000001)
Range is from -1 to +1

**Average Sentence Length** = the number of words / the number of sentences

Note :- Use the given paragraph.txt for calculations



```python
import re
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize ,sent_tokenize

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# Read the text document
with open('/content/drive/MyDrive/AI intern Assignment-20230705T154638Z-001/AI intern Assignment/paragraph.txt', 'r', encoding='utf-8') as file:
    text = file.read()

##<<---------------------Part 1-------------------->>

# Remove punctuation marks and convert to lowercase
text = re.sub(r'[^\w\s.]', '', text.lower())

# Set up NLTK resources
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = []

for token in (text.split()):
    if token not in stop_words and len(token) > 2:
        # Lemmatize the token
        lemmatized_token = lemmatizer.lemmatize(token)
        lemmatized_tokens.append(lemmatized_token)

# Count the frequency of each word
word_counts = Counter(lemmatized_tokens)

# Display the top 5 most frequent words
top_words = word_counts.most_common(5)
print("Top 5 most frequent words:")
for word, count in top_words:
    print(f"{word}: {count}")

##<<---------------------Part 1-------------------->>

##<<---------------------Part 2-------------------->>

with open('/content/drive/MyDrive/AI intern Assignment-20230705T154638Z-001/AI intern Assignment/positive-words.txt', 'r', encoding='utf-8') as file:
    positWord = file.read()
positWord= positWord.lower()
positWord= set(positWord.split())
# posit_present={}

dict(word_counts).keys()

pr = {key: value for key, value in word_counts.items() if key in positWord}
print()
print("Positive Dict-",pr)
ps= sum(pr.values())


with open('/content/drive/MyDrive/AI intern Assignment-20230705T154638Z-001/AI intern Assignment/negative-words.txt', 'r', encoding='latin-1') as file:
    negWord = file.read()
negtWord= negWord.lower()
negWord= set(negWord.split())

dict(word_counts).keys()

nr = {key: value for key, value in word_counts.items() if key in negWord}
print("Negative Dict-",nr)
ns= sum(nr.values())

# Calculate polarity score
polarity_score = ( ps- ns) / (ps + ns + 0.000001)
#print(polarity_score)


sl= len(sent_tokenize(text))

wl= len(text.split())
avgWordPerSen= wl/sl

# Display the calculated parameters
print(f"\nPositive Score: {ps}")
print(f"Negative Score: {ns}")
print(f"Polarity Score: {polarity_score:.4f}")
print(f"Average Sentence Length: {avgWordPerSen:.2f} words/sentence")
```
